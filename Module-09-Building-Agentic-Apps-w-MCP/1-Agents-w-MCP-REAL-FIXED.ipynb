# LAB 2: Connect to REAL MCP Servers + Tool Discovery (FIXED)

````python
# Global storage for MCP sessions and tool catalog
mcp_sessions = {}
mcp_contexts = {}
mcp_tool_catalog = {}  # NEW: Store discovered tools

async def connect_to_mcp_server(server_name: str, server_config: dict):
    """
    Connect to a REAL MCP server and keep the session alive.
    """
    command = server_config['command']
    args = server_config['args']
    env = server_config.get('env', {})
    
    # Merge environment variables
    full_env = os.environ.copy()
    full_env.update(env)
    
    server_params = StdioServerParameters(
        command=command,
        args=args,
        env=full_env
    )
    
    print(f"🔌 Starting {server_name} MCP server process...")
    
    # Start the MCP server as a subprocess
    context = stdio_client(server_params)
    read, write = await context.__aenter__()
    
    # Store the context manager for cleanup later
    mcp_contexts[server_name] = context
    
    # Create session
    session = ClientSession(read, write)
    await session.initialize()
    
    # Store session globally
    mcp_sessions[server_name] = session
    
    # List available tools
    response = await session.list_tools()
    
    print(f"\n📦 {server_name} MCP Server Tools:")
    print("─" * 60)
    
    # Store tools in catalog
    mcp_tool_catalog[server_name] = {}
    
    for tool_info in response.tools:
        tool_desc = tool_info.description[:80] + "..." if len(tool_info.description) > 80 else tool_info.description
        print(f"  🔧 {tool_info.name}")
        print(f"     {tool_desc}")
        
        # Store tool info
        mcp_tool_catalog[server_name][tool_info.name] = {
            'description': tool_info.description,
            'inputSchema': tool_info.inputSchema if hasattr(tool_info, 'inputSchema') else {}
        }
    
    print("")
    
    return session, response.tools

async def disconnect_all_mcp_servers():
    """Clean up all MCP server connections."""
    print("\n🔌 Disconnecting MCP servers...")
    for server_name, context in mcp_contexts.items():
        try:
            await context.__aexit__(None, None, None)
            print(f"  ✅ Disconnected {server_name}")
        except Exception as e:
            print(f"  ⚠️ Error disconnecting {server_name}: {e}")
    mcp_sessions.clear()
    mcp_contexts.clear()

print("✅ MCP connection functions ready (with tool catalog)")
````

````python
# Connect to MCP servers and discover tools
async def initialize_mcp_servers():
    """Initialize all MCP servers from config."""
    with open('/content/mcp_config.json') as f:
        config = json.load(f)
    
    print("🚀 Initializing REAL MCP servers...\n")
    
    all_tools = {}
    for server_name, server_config in config['mcpServers'].items():
        try:
            session, tools = await connect_to_mcp_server(server_name, server_config)
            all_tools[server_name] = tools
            print(f"✅ {server_name}: Connected with {len(tools)} tools\n")
        except Exception as e:
            print(f"❌ {server_name}: Failed - {str(e)}\n")
            import traceback
            print(traceback.format_exc())
    
    return all_tools

# Run initialization
print("⏳ Connecting to MCP servers (this may take 10-20 seconds)...\n")
available_tools = await initialize_mcp_servers()

print("\n" + "=" * 80)
print("🎉 All MCP servers are now running!")
print("=" * 80)

# Print tool summary
print("\n📋 DISCOVERED TOOL SUMMARY:")
print("=" * 80)
for server_name, tools_dict in mcp_tool_catalog.items():
    print(f"\n🔌 {server_name}: {len(tools_dict)} tools")
    for tool_name in tools_dict.keys():
        print(f"   • {tool_name}")

print("\n✅ Lab 2 Complete!")
````

# LAB 3: Gmail via REAL MCP Server (FIXED WITH DYNAMIC TOOL DISCOVERY)

````python
# Enhanced wrapper with better error handling and logging
async def call_mcp_tool_async(server_name: str, tool_name: str, arguments: dict):
    """
    Call a tool on a REAL MCP server with detailed logging.
    """
    session = mcp_sessions.get(server_name)
    if not session:
        raise ValueError(f"MCP server {server_name} not connected")
    
    print(f"\n🔧 MCP Tool Call:")
    print(f"   Server: {server_name}")
    print(f"   Tool: {tool_name}")
    print(f"   Arguments: {json.dumps(arguments, indent=2)}")
    
    result = await session.call_tool(tool_name, arguments)
    
    print(f"✅ MCP tool call successful!")
    
    return result

def call_mcp_tool_sync(server_name: str, tool_name: str, arguments: dict):
    """
    Synchronous wrapper for calling MCP tools (for use in LangChain).
    """
    loop = asyncio.get_event_loop()
    return loop.run_until_complete(call_mcp_tool_async(server_name, tool_name, arguments))

# Function to find the actual tool name dynamically
def find_tool_name(server_name: str, possible_names: list) -> str:
    """Find the actual tool name from a list of possibilities."""
    if server_name not in mcp_tool_catalog:
        print(f"⚠️ Warning: {server_name} not in tool catalog")
        return possible_names[0]  # Return first as fallback
    
    available_tools = mcp_tool_catalog[server_name].keys()
    
    for name in possible_names:
        if name in available_tools:
            print(f"✅ Found tool: {name}")
            return name
    
    print(f"⚠️ Warning: None of {possible_names} found in {server_name}")
    print(f"Available tools: {list(available_tools)}")
    return possible_names[0]  # Return first as fallback

# Discover actual Gmail tool names
print("🔍 Discovering Gmail tool names...")
GMAIL_SEND_TOOL = find_tool_name('google-workspace', [
    'send_gmail',
    'gmail_send',
    'send_email',
    'gmail_send_message',
    'send_message'
])

GMAIL_SEARCH_TOOL = find_tool_name('google-workspace', [
    'search_gmail',
    'gmail_search',
    'search_messages',
    'gmail_search_messages',
    'list_messages'
])

print(f"\n📧 Gmail Tools Configuration:")
print(f"   Send: {GMAIL_SEND_TOOL}")
print(f"   Search: {GMAIL_SEARCH_TOOL}")

# Create LangChain tool wrappers for Gmail MCP server
@tool
def send_gmail_via_real_mcp(to: str, subject: str, body: str) -> str:
    """
    Send email via REAL mcp-google server.
    
    Args:
        to: Recipient email address
        subject: Email subject line
        body: Email body text
    """
    try:
        print(f"\n📧 SENDING EMAIL VIA REAL MCP SERVER")
        print(f"═" * 60)
        
        result = call_mcp_tool_sync(
            server_name="google-workspace",
            tool_name=GMAIL_SEND_TOOL,
            arguments={
                "to": to,
                "subject": subject,
                "body": body
            }
        )
        
        print(f"\n" + "═" * 60)
        print(f"✅ EMAIL SENT SUCCESSFULLY!")
        print(f"═" * 60)
        print(f"From: {SENDER_EMAIL}")
        print(f"To: {to}")
        print(f"Subject: {subject}")
        print(f"MCP Server: google-workspace")
        print(f"Tool: {GMAIL_SEND_TOOL}")
        result_str = str(result.content[:200] if hasattr(result, 'content') else result[:200])
        print(f"Result: {result_str}...")
        print(f"═" * 60)
        
        return f"✅ Email sent via REAL MCP server! Check {to} inbox.")
    except Exception as e:
        import traceback
        error_details = traceback.format_exc()
        print(f"\n❌ ERROR SENDING EMAIL:")
        print(f"═" * 60)
        print(error_details)
        print(f"═" * 60)
        return f"❌ Failed to send email: {str(e)}"

@tool
def search_gmail_via_real_mcp(query: str, max_results: int = 10) -> str:
    """
    Search Gmail via REAL mcp-google server.
    
    Args:
        query: Search query
        max_results: Maximum number of results
    """
    try:
        result = call_mcp_tool_sync(
            server_name="google-workspace",
            tool_name=GMAIL_SEARCH_TOOL,
            arguments={
                "query": query,
                "max_results": max_results
            }
        )
        return str(result)
    except Exception as e:
        import traceback
        return f"❌ Search failed: {str(e)}\n{traceback.format_exc()}"

# Create agent with REAL MCP tools
gmail_tools = [send_gmail_via_real_mcp, search_gmail_via_real_mcp]
gmail_agent = create_agent(
    model="openai:gpt-4o",
    tools=gmail_tools,
    system_prompt=f"Gmail agent using REAL MCP server. Sender: {SENDER_EMAIL}"
)

print("\n✅ Gmail agent ready (using REAL MCP server)")
print(f"📧 Sender: {SENDER_EMAIL}")
print(f"🔌 Server: google-workspace (mcp-google)")
print(f"🔧 Tools: {GMAIL_SEND_TOOL}, {GMAIL_SEARCH_TOOL}")
````

# LAB 4: Calendar via REAL MCP Server (FIXED WITH DYNAMIC TOOL DISCOVERY)

````python
# Discover actual Calendar tool names
print("🔍 Discovering Calendar tool names...")
CALENDAR_CREATE_TOOL = find_tool_name('google-workspace', [
    'create_calendar_event',
    'calendar_create_event',
    'create_event',
    'calendar_create',
    'add_calendar_event',
    'add_event'
])

CALENDAR_LIST_TOOL = find_tool_name('google-workspace', [
    'list_calendar_events',
    'calendar_list_events',
    'list_events',
    'get_calendar_events',
    'get_events'
])

print(f"\n📅 Calendar Tools Configuration:")
print(f"   Create: {CALENDAR_CREATE_TOOL}")
print(f"   List: {CALENDAR_LIST_TOOL}")

# Create LangChain tool wrappers for Calendar MCP server
@tool
def create_calendar_event_via_real_mcp(
    summary: str,
    start_time: str,
    end_time: str,
    attendees: str,
    description: str = "",
    location: str = ""
) -> str:
    """
    Create Google Calendar event via REAL mcp-google server.
    
    Args:
        summary: Event title
        start_time: ISO format (2025-10-23T10:00:00+05:30)
        end_time: ISO format (2025-10-23T11:30:00+05:30)
        attendees: Comma-separated emails
        description: Event description
        location: Location or meeting link
    """
    try:
        print(f"\n📅 CREATING CALENDAR EVENT VIA REAL MCP SERVER")
        print(f"═" * 60)
        
        # Parse attendees
        attendee_list = [email.strip() for email in attendees.split(',')]
        
        result = call_mcp_tool_sync(
            server_name="google-workspace",
            tool_name=CALENDAR_CREATE_TOOL,
            arguments={
                "summary": summary,
                "start": {"dateTime": start_time, "timeZone": "Asia/Kolkata"},
                "end": {"dateTime": end_time, "timeZone": "Asia/Kolkata"},
                "attendees": attendee_list,
                "description": description,
                "location": location
            }
        )
        
        print(f"\n" + "═" * 60)
        print(f"✅ CALENDAR EVENT CREATED SUCCESSFULLY!")
        print(f"═" * 60)
        print(f"Event: {summary}")
        print(f"Start: {start_time}")
        print(f"End: {end_time}")
        print(f"Attendees: {', '.join(attendee_list)}")
        print(f"Location: {location}")
        print(f"MCP Server: google-workspace")
        print(f"Tool: {CALENDAR_CREATE_TOOL}")
        result_str = str(result)[:200]
        print(f"Result: {result_str}...")
        print(f"═" * 60)
        
        return f"✅ Calendar event created! Invites sent to {', '.join(attendee_list)}"
    except Exception as e:
        import traceback
        error_details = traceback.format_exc()
        print(f"\n❌ ERROR CREATING EVENT:")
        print(f"═" * 60)
        print(error_details)
        print(f"═" * 60)
        return f"❌ Failed to create event: {str(e)}"

@tool
def list_calendar_events_via_real_mcp(max_results: int = 10) -> str:
    """
    List upcoming calendar events via REAL mcp-google server.
    
    Args:
        max_results: Maximum number of events to return
    """
    try:
        now = datetime.utcnow().isoformat() + 'Z'
        result = call_mcp_tool_sync(
            server_name="google-workspace",
            tool_name=CALENDAR_LIST_TOOL,
            arguments={
                "time_min": now,
                "max_results": max_results
            }
        )
        return str(result)
    except Exception as e:
        import traceback
        return f"❌ Failed to list events: {str(e)}\n{traceback.format_exc()}"

# Create agent with REAL MCP tools
calendar_tools = [create_calendar_event_via_real_mcp, list_calendar_events_via_real_mcp]
calendar_agent = create_agent(
    model="openai:gpt-4o",
    tools=calendar_tools,
    system_prompt="Calendar agent using REAL MCP server."
)

print("\n✅ Calendar agent ready (using REAL MCP server)")
print(f"🔌 Server: google-workspace (mcp-google)")
print(f"🔧 Tools: {CALENDAR_CREATE_TOOL}, {CALENDAR_LIST_TOOL}")
````
