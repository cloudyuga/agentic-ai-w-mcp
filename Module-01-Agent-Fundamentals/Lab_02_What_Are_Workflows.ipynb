```
# %% [markdown]
# # Lab 2: What Are Workflows?
#
# **Goal:** Understand the concept of a structured, repeatable workflow, which is the foundation upon which we will build our agent.
#
# ---

# %% [markdown]
# ## Part A: The Manual Recruitment Process
#
# Before we can automate a process, we must first understand it. Below is a typical, high-level workflow for recruiting a new candidate. It's a structured, step-by-step process.

# %%
# A visual representation of a manual workflow
manual_process = """
Step 1: Receive Resume (from Email or Job Portal)
  |
  V
Step 2: HR Screens Resume (Manual review against job description)
  |
  V
Step 3: Decision: Is the candidate qualified?
  |
  +-- YES --> Step 4: Schedule Interview
  |
  +-- NO ---> Step 7: Send Rejection Email
  |
  V
Step 4: Conduct Interview (Technical & Cultural)
  |
  V
Step 5: Collect Interview Feedback
  |
  V
Step 6: Make Final Decision (Hire/Reject)
  |
  +-- HIRE --> Step 7: Send Offer Email
  |
  +-- REJECT -> Step 7: Send Rejection Email
"""

print("A Typical Manual Recruitment Workflow:")
print(manual_process)

# %% [markdown]
# **Exercise:** Think about the hiring process at your own organization or a company you know. Can you map it out in a similar step-by-step fashion? Does it follow a predictable sequence?
#
# ---

# %% [markdown]
# ## Part B: Key Components of a Workflow
#
# Every workflow, regardless of the domain, is composed of a few key elements. Let's identify them in our recruitment example.
#
# 1.  **Input:** The starting material.
#     * *Example:* The candidate's resume and application form.
#
# 2.  **Process Steps:** The actions taken.
#     * *Example:* Screening the resume, conducting the interview, making a final decision.
#
# 3.  **Decision Points:** Forks in the road where the process can go in different directions based on a condition.
#     * *Example:* Is the resume score high enough? Was the interview feedback positive?
#
# 4.  **Output:** The final result or outcome of the workflow.
#     * *Example:* A hire/reject decision, an offer letter sent, a rejection email sent.
#
# 5.  **Parallel Tasks (Optional):** Tasks that can happen at the same time.
#     * *Example:* While scheduling an interview, HR could also be running a background check.
#
# 6.  **Human Involvement:** Points where a human must intervene.
#     * *Example:* The interview itself, the final hiring decision.
#
# ---

# %% [markdown]
# ## Part C: Characteristics of a Workflow
#
# Workflows are best suited for tasks that have the following characteristics.
#
# * **Predictable:** The steps are known in advance and don't change unexpectedly.
# * **Structured:** There is a clear and logical sequence of events.
# * **Rule-based:** The logic is based on clear "if-then" conditions.
# * **Repeatable:** The process is performed consistently for every new input.
#
# **Exercise:** Look at the manual recruitment process again. Can you identify where each of these characteristics applies? For example, the decision to schedule an interview is **rule-based** (e.g., IF score > 80, THEN schedule).
```
